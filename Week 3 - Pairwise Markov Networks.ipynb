## Undirected Networks are generally called Markov Networks 

The easiest case are **pairwise Markov networks**

```python
from graphviz import Graph
g = Graph('G')

g.edge('Alice','Bob')
g.edge('Bob', 'Charles')
g.edge('Alice', 'Debbie')
g.edge('Debbie', 'Charles')
g
```

How do you parameterize such a graph? We can't condition! 

We can use factors: compatibility functions, soft constraints, etc.

so e.g. $\phi_1$ for A and B: 

| a | b | value |
|---|---|---|
| a0 | b0 | 30 | 
| a1 | b0 | 0 |
| a0 | b1 | 0 |
| a1 | b1 | 10 |

for B, C:

| b | c | value |
|---|---|---|
| b0 | c0 | 100 | 
| b1 | c0 | 1 |
| b0 | c1 | 1 |
| b1 | b1 | 10 |

etc.

We will use products of factors: 

$$\large \tilde{P}(A,B,C,D) = \phi_1 (A,B) \times \phi_2 (B, C) \times \phi_3 (A, D) \times \phi_4 (A, D)$$

Note that this is not a real probability distribution, since there's no guarantee that it lies between 0 and 1. This is denoted by the tilde $\rightarrow$ unnormalized measure

Thus $P(A,B,C,D) = \frac{1}{Z} \tilde{P}(A,B,C,D)$, where $Z$ is the partition function, here the sum of all the unnormalized probabilities.



### Potentials in Markov Networks are unintuitive

The factor product enforces "logic". E.g. if A and D like to agree, and B and C like to agree, but C and D like to disagree, then we can't have that A and B have a high probability of agreeing. 

**There isn't a natural mapping between the probability distributions and the factors**

### Definition of Pairwise Markov Networks

An undirected graph whose nodes are $X_1,\dotsc , X_n$ and each edge $X_i - X_j$ is associated with a factor (potential) $\phi_{ij} (X_i - X_j)$
