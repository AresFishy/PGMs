## Local Structure

This is necessary when we have continuous variables

e.g. a sensor measuring temperature: $S \sim N(T ; \sigma_s^2)$

This is called a **linear Gaussian**


```python
from graphviz import Digraph

g = Digraph('G', engine='dot')
g.edge('Temperature', 'Sensor')
g.edge('Temperature', "Temperature'")
g.edge('Outside Temp', "Temperature'")
g
```

We can include Temperature' at the next time step, given by e.g. a diffusion model:

$$ T' \sim N(\alpha T + (1-\alpha)O \ ; \ \sigma_T^2)$$

We can have different parameters for the Gaussian describing T' if we have more variables, e.g. door open or closed. This would be called a **conditional Gaussian**

###  Linear Gaussian

$$Y \sim N(w_0 + \sum w_i X_i \ ; \ \sigma^2)$$

* The mean is a linear function of the parents
* The variance is **independent** of the parents

### Conditional Linear Gaussian

$$Y \sim N(w_{a0} + \sum w_{ai} X_i \ ; \ \sigma_a^2)$$

### Nonlinear Gaussians

We can capture interesting things with these: e.g. an estimation of the position of an obstacle. This is (hopefully) a Gaussian centered around the true position, but preceded by a uniform probability region and followed by near-zero probability: the model "knows" that information following the obstacle may interfere with the sensor.
